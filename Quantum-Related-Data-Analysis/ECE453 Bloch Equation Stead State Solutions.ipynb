## Coupled Bloch Equation steady-state and non-stead-state comparisons

```python
import numpy as np
import matplotlib.pyplot as plt


gamma_sp = 1.0
gamma_perp = 1.0       
Omega_list = [0.01, 0.1, 1.0, 10.0]

deltaT = 0.001
ti = 0.0
tf = 10.0
time = np.linspace(ti, tf, int((tf - ti)/deltaT) + 1)
plt.figure(figsize=(10,8))

Omega_R = .01
p = np.zeros(len(time), dtype=complex)
d = np.zeros(len(time))
p[0] = 0.0 + 0.0j
d[0] = -1.0

for i in range(len(time)-1):

    pr = p[i].real
    pi = p[i].imag
    di = d[i]
    dpr_dt = -gamma_perp * pr
    dpi_dt = -gamma_perp * pi - Omega_R * di
    dd_dt  = -gamma_perp*2 * (1 + di) + 4*Omega_R * pi

    p[i+1] = pr + dpr_dt*deltaT + 1j*(pi + dpi_dt*deltaT)
    d[i+1] = di + dd_dt * deltaT

pr = p.real
pi = p.imag

plt.plot(time, pr, label=f"Re[p], ratio={Omega_R}")
plt.plot(time, pi, '--', label=f"Im[p], ratio={Omega_R}")
plt.plot(time, d, ':', label=f"d, ratio={Omega_R}")

plt.xlabel("time (units of 1/gamma_sp)")
plt.ylabel("Amplitude (scaled by 1/P12)")
plt.title("Amplitude of p_r, p_i, and d for Ratio = 0.01")
plt.legend()
plt.grid(True)
plt.show()
```
